```python
import json

def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
        
    return data
```

```python
data=load_data("data.json")
print(data)
```

```python
#clean and structure the data 
def clean_data(data):
    text_to_num= {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    for user in data:
        #cleaning rating
        raw_rating=user["rating"].strip().lower()
        if raw_rating in text_to_num:
            raw_rating=text_to_num[raw_rating]
        #print(raw_rating)
        user["rating"]= raw_rating

        #Handle Missing Age
        raw_age=user.get("age")
        if 'raw_age'==None:
            user['age']=None

        #Duplication user
        if user['name'].strip() in unique_users:
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

```python
clean_data(data)
```

```python
#Get meaningfull insights
def get_insights(data):
    #avg rating
    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])

    print(f"avg rating = {tot_rating/len(data)}")

    #percentage of user with poor rating
    poor_rating=0
    for user in data:
        if (float(user["rating"] ) < 3):
            poor_rating+=1
              
    print(f"% of user with poor rating = {poor_rating/len(data) * 100}%")
```

```python
get_insights(data)
```

```python
def get_recommandation(data):
    recommandation=[]

    for user in data:
        curr_recomm={} 
        curr_recomm["name"]= user["name"]

        if (float(user["rating"] )>= 4):
            curr_recomm["brand"] ="Apple"
        else:
            curr_recomm["brand"] ="Samsung"

            recommandation.append(curr_recomm)
    return recommandation
```

```python
get_recommandation(data)
```
