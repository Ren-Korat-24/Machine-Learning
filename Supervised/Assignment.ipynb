```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))
scores= np.random.normal(loc=70, scale=10, size=200)

means=scores.mean()
plt.axvline(means,color="red",linestyle="--",label=f"Means={means:.2f}")

plt.hist(scores,bins=20,edgecolor="black")
plt.title("Distribution of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
import seaborn as sns
penguins=sns.load_dataset("penguins")

sns.scatterplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)

plt.title("flipper_length_mm vs body_mass_g")
plt.show()
```

```python
# import seaborn as sns
tips=sns.load_dataset("tips")

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title("Total bill by Day and Sex")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

pivot = tips.pivot_table(
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean"
)

pivot
```

```python
sns.heatmap(
    pivot,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu"
)

plt.title("Average Tip by Day and Time")
plt.xlabel("Time")
plt.ylabel("Day")
plt.show()
```
